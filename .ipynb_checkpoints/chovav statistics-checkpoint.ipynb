```python
from Classes.Torah import Torah
import pandas as pd
from collections import Counter, defaultdict
import matplotlib.pyplot as plt
import numpy
```

```python
torah_roots = Torah()
torah_roots.load("torah_roots_without_nikud.pkl")

torah = Torah()
torah.load("torah.pkl")
```

```python
roots_global_counter = Counter()
words_global_counter = Counter()

book_roots_counters = defaultdict(Counter)
book_words_counters = defaultdict(Counter)

teuda_roots_counters = defaultdict(Counter)
teuda_words_counters = defaultdict(Counter)
```

```python
for book in torah.books:
    for pasuk in book.psukim:
        words_global_counter.update(pasuk.words())
        book_words_counters[book.book_name].update(pasuk.words())

for book in torah_roots.books:
    for pasuk in book.psukim:
        roots_global_counter.update(pasuk.constituency_tree.roots)
        book_roots_counters[book.book_name].update(pasuk.constituency_tree.roots)

for teuda in torah.teudot:
    for pasuk in teuda.psukim:
        teuda_words_counters[teuda.teuda_name].update(pasuk.words())

for teuda in torah_roots.teudot:
    for pasuk in teuda.psukim:
        teuda_roots_counters[teuda.teuda_name].update(pasuk.constituency_tree.roots)
```

```python
with open("heb-stopwords.txt", "r") as fp:
    stopwords = fp.read()
stopwords = stopwords.split("\n")
```

```python
for stopw in stopwords:
    for c in book_roots_counters.values():
        if stopw in c:
            c.pop(stopw)
    for c in book_words_counters.values():
        if stopw in c:
            c.pop(stopw)

    for c in teuda_roots_counters.values():
        if stopw in c:
            c.pop(stopw)
    for c in teuda_words_counters.values():
        if stopw in c:
            c.pop(stopw)
    
    if stopw in roots_global_counter:
        roots_global_counter.pop(stopw)
    if stopw in words_global_counter:
        words_global_counter.pop(stopw)
```

```python
D_counter = teuda_roots_counters['D'].most_common()
```

```python
rankers = {
    t: [k for k,v in c.most_common()]
    for t, c in teuda_roots_counters.items()
}
```

```python
for w, _ in roots_global_counter.most_common(20):
    # print(w)
    if w in rankers['R']:
        i = rankers['R'].index(w) + 1
        if i <= 20:
            print(i)
        else:
            print(24)
    else:
        print(24)
```

```python
book_lengths = {
    book.book_name: len(book.psukim)
    for book in torah.books
}
teuda_lengths = {
    teuda.teuda_name: len(teuda.psukim)
    for teuda in torah.teudot
}

def teuda_absolute_value(val):
    sizes = numpy.array(list(teuda_lengths.values()))
    a  = int(numpy.round(val/100*sizes.sum(), 0))
    return f"{a}\n{val:.1f}%"

def book_absolute_value(val):
    sizes = numpy.array(list(book_lengths.values()))
    a  = int(numpy.round(val/100*sizes.sum(), 0))
    return f"{a}\n{val:.1f}%"
 
```

```python
plt.figure(figsize=(8, 8))
plt.pie(teuda_lengths.values(), labels=teuda_lengths.keys(), autopct=teuda_absolute_value, startangle=140, 
        wedgeprops={'edgecolor': 'black', 'width': 0.5}, pctdistance=0.75, textprops={'fontsize': 16})
# plt.show()
plt.savefig('teudot.png')
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    book_lengths.values(), 
    labels=book_lengths.keys(), 
    autopct=book_absolute_value, 
    startangle=140, 
    wedgeprops={'edgecolor': 'black', 'width': 0.5}, 
    pctdistance=0.75,
    textprops={'fontsize': 16}
)
# plt.show()
plt.savefig('books.png')
```

```python
book_words = {
    book.book_name: sum([p.word_count() for p in book.psukim])
    for book in torah.books
}
teuda_words = {
    teuda.teuda_name: sum([p.word_count() for p in teuda.psukim])
    for teuda in torah.teudot
}

def teuda_absolute_value(val):
    sizes = numpy.array(list(teuda_words.values()))
    a  = int(numpy.round(val/100*sizes.sum(), 0))
    return f"{a}\n{val:.1f}%"

def book_absolute_value(val):
    sizes = numpy.array(list(book_words.values()))
    a  = int(numpy.round(val/100*sizes.sum(), 0))
    return f"{a}\n{val:.1f}%"
 
```

```python
plt.figure(figsize=(8, 8))
plt.pie(book_words.values(), labels=book_words.keys(), autopct=book_absolute_value, startangle=140, 
        wedgeprops={'edgecolor': 'black', 'width': 0.5}, pctdistance=0.75, textprops={'fontsize': 16})
# plt.show()
plt.savefig('book_words.png')
```

```python
plt.figure(figsize=(8, 8))
plt.pie(teuda_words.values(), labels=teuda_words.keys(), autopct=teuda_absolute_value, startangle=140, 
        wedgeprops={'edgecolor': 'black', 'width': 0.5}, pctdistance=0.75, textprops={'fontsize': 16})
# plt.show()
plt.savefig('teuda_words.png')
```

```python
book_distinct_words = defaultdict(set)
teuda_distinct_words = defaultdict(set)
```

```python
for teuda in torah_roots.teudot:
    for pasuk in teuda.psukim:
        for w in pasuk.constituency_tree.roots:
            teuda_distinct_words[teuda.teuda_name].add(w)

for book in torah_roots.books:
    for pasuk in book.psukim:
        for w in pasuk.constituency_tree.roots:
            book_distinct_words[book.book_name].add(w)
```

```python
book_distinct_words_count = {
    book_name: len(words)
    for book_name, words in book_distinct_words.items()
}
teuda_distinct_words_count = {
    teuda_name: len(words)
    for teuda_name, words in teuda_distinct_words.items()
}
```

```python
def teuda_absolute_value(val):
    sizes = numpy.array(list(teuda_distinct_words_count.values()))
    a  = int(numpy.round(val/100*sizes.sum(), 0))
    return f"{a}\n{val:.1f}%"

def book_absolute_value(val):
    sizes = numpy.array(list(book_distinct_words_count.values()))
    a  = int(numpy.round(val/100*sizes.sum(), 0))
    return f"{a}\n{val:.1f}%"
 
```

```python
plt.figure(figsize=(8, 8))
plt.pie(book_distinct_words_count.values(), labels=book_distinct_words_count.keys(), autopct=book_absolute_value, startangle=140, 
        wedgeprops={'edgecolor': 'black', 'width': 0.5}, pctdistance=0.75, textprops={'fontsize': 16})
# plt.show()
plt.savefig('book_distinct_words_count.png')
```

```python
plt.figure(figsize=(8, 8))
plt.pie(teuda_distinct_words_count.values(), labels=teuda_distinct_words_count.keys(), autopct=teuda_absolute_value, startangle=140, 
        wedgeprops={'edgecolor': 'black', 'width': 0.5}, pctdistance=0.75, textprops={'fontsize': 20})
# plt.show()
plt.savefig('teuda_distinct_words_count.png')
```
